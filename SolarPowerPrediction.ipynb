```python

import matplotlib.pyplot as plt
import seaborn as sns


```

```python

# Histogram: Distribution of temperature
plt.figure(figsize=(10, 6))
sns.histplot(df['temperature_2_m_above_gnd'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Temperature (2m above ground)')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box Plot: Total precipitation by cloud cover levels
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='total_cloud_cover_sfc', y='total_precipitation_sfc', palette='Set3')
plt.title('Total Precipitation vs Cloud Cover')
plt.xlabel('Total Cloud Cover (%)')
plt.ylabel('Total Precipitation (mm)')
plt.show()
```

```python
# Heatmap: Correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix of Features')
plt.show()
```

```python
# Bar Plot: Average generated power by cloud cover levels
avg_power_by_cloud_cover = df.groupby('total_cloud_cover_sfc')['generated_power_kw'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=avg_power_by_cloud_cover, x='total_cloud_cover_sfc', y='generated_power_kw', palette='viridis')
plt.title('Average Generated Power by Cloud Cover Levels')
plt.xlabel('Total Cloud Cover (%)')
plt.ylabel('Average Generated Power (kW)')
plt.show()
```
